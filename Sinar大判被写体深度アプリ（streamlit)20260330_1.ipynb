```python
def sinar_p25_calculator():
    print("--- дјҠи—ӨгҒ•гӮ“е°Ӯз”Ё: Sinar P & P25 ж’®еҪұиЁҲз®—ж©ҹ ---")

    # еҹәжң¬иЁӯе®ҡ (Phase One P25)
    pixel_size_um = 9.0  # P25гҒ®з”»зҙ гғ”гғғгғҒ
    focal_length_mm = 90  # дҪҝз”Ёгғ¬гғігӮә (Fujinon 90mmзӯү)

    # иЁұе®№йҢҜд№ұеҶҶгҒ®е®ҡзҫ© (гғҮгӮёгӮҝгғ«гғҗгғғгӮҜз”ЁгҒ«еҺігҒ—гӮҒгҒ«иЁӯе®ҡ)
    # йҖҡеёёгҒ®4x5гғ•гӮЈгғ«гғ (0.1mm)гҒ«еҜҫгҒ—гҖҒP25гҒ§гҒҜзҙ„0.03mmзЁӢеәҰгҒҢзҗҶжғі
    coc_mm = 0.03

    try:
        # е…ҘеҠӣгӮ»гӮҜгӮ·гғ§гғі (гғ¬гғјгӮ¶гғји·қйӣўиЁҲгҒ®ж•°еҖӨгӮ’е…ҘеҠӣ)
        dist_near = float(input("иҝ‘зӮ№гҒҫгҒ§гҒ®и·қйӣў (m): "))
        dist_far = float(input("йҒ зӮ№гҒҫгҒ§гҒ®и·қйӣў (m): "))

        if dist_near >= dist_far:
            print("гӮЁгғ©гғј: йҒ зӮ№гҒҜиҝ‘зӮ№гӮҲгӮҠйҒ гҒ„еҝ…иҰҒгҒҢгҒӮгӮҠгҒҫгҒҷгҖӮ")
            return

        # 1. жңҖйҒ©гҒӘгғ”гғігғҲеҗҲгӮҸгҒӣи·қйӣў (иӘҝе’Ңе№іеқҮ)
        focus_dist = (2 * dist_near * dist_far) / (dist_near + dist_far)

        # 2. еҝ…иҰҒзөһгӮҠеҖӨ (FеҖӨ) гҒ®иЁҲз®—
        # иЁҲз®—ејҸ: F = (f^2 * О”D) / (2 * CoC * D1 * D2)
        delta_d = dist_far - dist_near
        required_f = (focal_length_mm**2 * delta_d * 1000) / (2 * coc_mm * (dist_near * 1000) * (dist_far * 1000))

        # 3. еӣһжҠҳйҷҗз•ҢгҒ®гғҒгӮ§гғғгӮҜ
        # дёҖиҲ¬зҡ„гҒ«йқ’иүІе…ү(550nm)еҹәжә–гҒ§гҖҒгӮЁгӮўгғӘгғјгғҮгӮЈгӮ№гӮҜеҫ„гҒҢз”»зҙ гғ”гғғгғҒгҒ®2еҖҚгӮ’и¶…гҒҲгӮӢгҒЁгғңгӮұгҒҢзӣ®з«ӢгҒӨ
        diffraction_limit_f = pixel_size_um / 1.5

        print("\n" + "="*40)
        print(f"гҖҗиЁҲз®—зөҗжһңгҖ‘")
        print(f"в–  жҺЁеҘЁгғ”гғігғҲдҪҚзҪ®: {focus_dist:.3f} m")
        print(f"в–  зҗҶи«–дёҠгҒ®еҝ…иҰҒзөһгӮҠ: F {required_f:.1f}")

        print("\nгҖҗзҸҫе ҙгҒ§гҒ®гӮўгӮҜгӮ·гғ§гғігҖ‘")
        if required_f > 16:
            print(f"вҡ пёҸ иӯҰе‘Ҡ: F{required_f:.1f}гҒҜзөһгӮҠгҒҷгҒҺгҒ§гҒҷ(еӣһжҠҳгғңгӮұгҒҢзҷәз”ҹгҒ—гҒҫгҒҷ)гҖӮ")
            print("   вҶ’ Sinar PгҒ®гӮўгӮӘгғӘпјҲгғҶгӮЈгғ«гғҲпјүгӮ’дҪҝгҒ„гҖҒгғ”гғігғҲйқўгӮ’еҖ’гҒ—гҒҰгҒҸгҒ гҒ•гҒ„гҖӮ")
        elif required_f < 5.6:
            print(f"вң… F8гҖң11гҒ§еҚҒеҲҶгҒ«гғ‘гғігғ•гӮ©гғјгӮ«гӮ№гҒҢеҫ—гӮүгӮҢгҒҫгҒҷгҖӮ")
        else:
            print(f"рҹ‘ү гғ¬гғігӮәгӮ’ F {math.ceil(required_f)} зЁӢеәҰгҒҫгҒ§зөһгҒЈгҒҰгҒҸгҒ гҒ•гҒ„гҖӮ")

        print(f"\nвҖ»P25гҒ®еӣһжҠҳйҷҗз•Ңзӣ®е®ү: F {diffraction_limit_f:.1f}")
        print("="*40)

    except ValueError:
        print("ж•°еҖӨгӮ’е…ҘеҠӣгҒ—гҒҰгҒҸгҒ гҒ•гҒ„гҖӮ")

sinar_p25_calculator()
```
